## Merge ratings and books on 'ISBN' 

```python
ratings_with_name = ratings.merge(books,on='ISBN')
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200

digik_users = x[x].index
ratings_with_name
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(digik_users)]

y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]

df = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')

df.fillna(0,inplace=True)


df
```

## Recommendation System

```python
# Get input from user for book title
user_input = input("Enter a book title to get recommendations: ")

# Filter ratings for the entered book title
filtered_ratings = ratings_with_name[ratings_with_name['Book-Title'].str.contains(user_input, case=False)]
# Calculate average rating for each book
book_ratings_avg = filtered_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Sort books by average rating in descending order
recommended_books = book_ratings_avg.sort_values(by='Book-Rating', ascending=False)

# Print recommended books
print(f"Recommended books related to '{user_input}':")
for index, row in recommended_books.iterrows():
    print(f"{row['Book-Title']}: Average Rating {row['Book-Rating']:.2f}")
    
```
